**pyplot**

pyplot provides a procedural interface to the matplotlib object-oriented plotting library. It is modeled closely after Matlab. Therefore, the majority of plotting commands in pyplot have Matlab analogs with similar arguments. Important commands are explained with interactive examples.

```python
import matplotlib.pyplot as plt
import numpy as np
```

`X` is now a numpy array with 256 values ranging from - to + (closed interval). C is the cosine (array with 256 values) and S is the sine.

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)


# Array columns C, S
M = np.stack((C,S), axis=1)

# Each column of the 2D array will comprise a separete set of data
plt.plot(X,M)
plt.legend(['Cos', 'Sin'], loc='upper left')
```

**Basic configuration**

```python
# Create a new figure with a size of width=8 and height=6 inches and resolution of 80 dots per inch (DPI)
plt.figure(figsize=(8, 6), dpi=80)

# Add a subplot to the figure (1 row, 1 column, 1st plot)
# This is useful when creating multiple plots in a grid layout
plt.subplot(1, 1, 1)

# Generate 256 evenly spaced values from - to 
X = np.linspace(-np.pi, np.pi, 256)

# Compute the cosine and sine values for each X
C, S = np.cos(X), np.sin(X)

# Plot the cosine curve:
# - X values on the x-axis
# - C (cosine of X) values on the y-axis
# - Blue continuous line, 1 pixel wide
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")

# Plot the sine curve:
# - X values on the x-axis
# - S (sine of X) values on the y-axis
# - Green dots-line, 3 pixel wide
plt.plot(X, S, color="green", linewidth=3.0, linestyle=":") # Style's types--->   --     -     :     ''    -

# Set the limits of the x-axis from -4 to 4 (wider than - to  for visual padding)
plt.xlim(-4.0, 4.0)

# Define tick marks along the x-axis at 9 evenly spaced intervals between -4 and 4
plt.xticks(np.linspace(-4, 4, 9))

# Set the limits of the y-axis from -1 to 1 (because sine and cosine values fall in this range)
plt.ylim(-1.0, 1.0)

# Define tick marks along the y-axis at 5 evenly spaced intervals between -1 and 1
plt.yticks(np.linspace(-1, 1, 5))

# Save the plot as a PNG file with 72 DPI (this line is commented out)
plt.savefig("exercise_2.png", dpi=72)

plt.title("Cosine and Sine Functions")           # Title of the plot
plt.xlabel("Angle [radians]")                    # X-axis label
plt.ylabel("Function value")                     # Y-axis label
plt.legend(["Cosine", "Sine"], loc="upper right")# Legend for the plotted curves
plt.grid(True)                                   # Enable grid lines for better readability

# Display the figure window with the plotted curves
plt.show()
```

**Setting tick labels**


Ticks are now properly placed but their label is not very explicit. We could guess that 3.142 is  but it would be better to make it explicit. **Moving spines**

Spines are the lines connecting the axis tick marks and noting the boundaries of the data area. They can be placed at arbitrary positions and until now, they were on the border of the axis. Well change that since we want to have them in the middle. Since there are four of them (top/bottom/left/right), well discard the top and right by setting their color to none and well move the bottom and left ones to coordinate 0 in data space coordinates.

```python
# Create a new figure window with specific dimensions (10 inches wide, 6 inches tall)
# and a resolution of 80 dots per inch. This controls the size and clarity of the plot.
plt.figure(figsize=(10, 6), dpi=80)

# Plot the cosine curve using array X (x-values) and C (cosine values),
# in blue color with a dashed line style and a thickness of 2.5 points.
# This makes the cosine function visually distinct.
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="--")

# Plot the sine curve using array X (x-values) and S (sine values),
# in red color with a dash-dot line style and a thicker line width.
# Differentiating sine from cosine visually with both color and line style.
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-.")

# Set specific tick positions on the x-axis at multiples of 
# and label them using LaTeX formatting to display  symbols nicely.
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Set the tick marks for the y-axis at -1, 0, and +1
# and use LaTeX formatting for better visual presentation.
plt.yticks([-1, 0, +1],
           [r'$-1$', r'$0$', r'$+1$'])

# Get the current Axes instance, which allows for detailed customization
# of the coordinate axes, ticks, and spines (the borders around the plot area).
ax = plt.gca()  # gca = get current axis

# Hide the right and top spines (borders) of the plot
# to make the plot cleaner and focus only on the main axes.
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Ensure that the x-axis ticks appear only at the bottom
# and move the bottom spine (x-axis line) to y=0 so that the axis crosses the origin.
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))

# Ensure that the y-axis ticks appear only on the left
# and move the left spine (y-axis line) to x=0 to center the axis on the origin.
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

# Render the complete plot on the screen
plt.show()
```

**Annotate some points**

Lets annotate some interesting points using the annotate command. We chose the 2/ 3 value and we want to annotate both the sine and the cosine. Well first draw a marker on the curve as well as a straight dotted line. Then, well use the annotate command to display some text with an arrow.

```python
# Define the point of interest: t = 2/3 radians
# This is where we will highlight the sine and cosine values
t = 2 * np.pi / 3

# Draw a vertical dashed line from y = 0 up to cos(t)
# This visually connects the angle t on the x-axis to its cosine value
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle="--")

# Mark the point (t, cos(t)) with a blue dot of size 50
plt.scatter([t], [np.cos(t)], s=50, color='blue')

# Add a label to the cosine point with LaTeX formatting
# The annotation has an arrow pointing to the cosine value at t
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',            # Location of the point being annotated
             xytext=(-90, -50), textcoords='offset points', # Position of the text relative to the point
             fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))  # Arrow style

# Draw a vertical dashed line from y = 0 up to sin(t)
# This shows the vertical distance representing sine at angle t
plt.plot([t, t], [0, np.sin(t)], color='red', linewidth=2.5, linestyle="--")

# Mark the point (t, sin(t)) with a red dot of size 50
plt.scatter([t], [np.sin(t)], s=50, color='red')

# Annotate the sine point similarly with LaTeX and an arrow
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points',
             fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

# Plot the full cosine curve in solid blue
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")

# Plot the full sine curve in solid red
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")

# Add a legend to help identify the sine and cosine curves
plt.legend(loc='upper left')

# Show the final annotated plot
plt.show()
```

**Figures, Subplots, Axes and Ticks**

A figure in matplotlib means the whole window in the user interface. Within this figure there can be subplots.

So far we have used implicit figure and axes creation. This is handy for fast plots. We can have more control over the display using figure, subplot, and axes explicitly. While subplot positions the plots in a regular grid, axes allows free placement within the figure. Both can be useful depending on your intention. Weve already worked with figures and subplots without explicitly calling them. When we call plot, matplotlib calls gca() to get the current axes and gca in turn calls gcf() to get the current figure. If there is none it calls figure() to make one, strictly speaking, to make a subplot(). Lets look at the details.

**Ticks**

Major and minor ticks can be located and formatted independently from each other. Per default minor ticks are not shown, i.e. there is only an empty list for them because it is as NullLocator



**Tick Locators**

Tick locators control the positions of the ticks. They are set as follows:

```python
import matplotlib.ticker as ticker  # Needed for custom locators

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data
plt.plot(x, y)

# Get current Axes object
ax = plt.gca()

# Define locator string (can be changed dynamically)
locator = "ticker.MultipleLocator(2)"  # Show a major tick every 2 units on the x-axis

# Set the x-axis major tick locator using eval
ax.xaxis.set_major_locator(eval(locator))

# Add grid for visibility
ax.grid(True)

# Display the plot
plt.show()
```

`ticker.MultipleLocator(2)` sets ticks at 0, 2, 4, ..., etc.

`eval(locator)` allows you to define the locator as a string and evaluate it at runtime. This could be useful if you get the locator as user input or from a config.

**Locators for tick on axis**

An example demoing different locators to position ticks on axis for matplotlib.

```python
# Define a helper function to set up a minimalist horizontal axis for tick demonstration
def tickline():
    # Set the limits of the plot
    plt.xlim(0, 10)         # x-axis range from 0 to 10
    plt.ylim(-1, 1)         # y-axis range from -1 to 1 (invisible)
    plt.yticks([])          # Remove y-axis ticks

    # Get current Axes object
    ax = plt.gca()

    # Hide the right, left, and top spines (borders)
    ax.spines["right"].set_color("none")
    ax.spines["left"].set_color("none")
    ax.spines["top"].set_color("none")

    # Move the bottom spine (x-axis) to y = 0
    ax.xaxis.set_ticks_position("bottom")
    ax.spines["bottom"].set_position(("data", 0))

    # Hide y-axis ticks
    ax.yaxis.set_ticks_position("none")

    # Add minor ticks every 0.1 (for context)
    ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.1))

    # Plot a flat horizontal line across the x-axis
    ax.plot(np.arange(11), np.zeros(11))

    return ax  # Return the configured Axes object

# Define a list of different tick locators as strings
locators = [
    "ticker.NullLocator()",               # No ticks
    "ticker.MultipleLocator(1.0)",        # Ticks every 1.0 unit
    "ticker.FixedLocator([0, 2, 8, 9, 10])", # Ticks only at specified fixed positions
    "ticker.IndexLocator(3, 1)",          # Ticks at every 3rd index, offset by 1
    "ticker.LinearLocator(5)",            # Evenly space 5 ticks across the axis
    "ticker.LogLocator(2, [1.0])",        # Logarithmic scale ticks (base 2)
    "ticker.AutoLocator()",               # Automatically chosen tick locations
]

n_locators = len(locators)  # Total number of locators to demonstrate

# Set up figure size and resolution
size = 512, 40 * n_locators           # Width: 512 px, Height: 40 px per locator
dpi = 72.0                            # Dots per inch (for sizing)
figsize = size[0] / dpi, size[1] / dpi  # Convert pixel size to inches
fig = plt.figure(figsize=figsize, dpi=dpi)  # Create the figure
fig.patch.set_alpha(0)                # Make the figure background transparent


# Loop through each locator type and plot its effect
for i, locator in enumerate(locators):
    # Create a subplot for each locator demonstration
    plt.subplot(n_locators, 1, i + 1)

    # Create a minimal axis line with no y-ticks and a horizontal baseline
    ax = tickline()

    # Evaluate the locator string and apply it as the major tick locator on the x-axis
    ax.xaxis.set_major_locator(eval(locator))

    # Annotate the subplot with the name of the locator being used
    plt.text(5, 0.3, locator[7:], ha="center")  # Centered text showing the locator name

# Adjust subplot spacing to minimize padding between plots
plt.subplots_adjust(bottom=0.01, top=0.99, left=0.01, right=0.99)

# Display the composed figure with all locator types
plt.show()
```

**Regular Plots**

```python
# Number of points used to sample the function
n = 256

# Generate 256 values from - to 
X = np.linspace(-np.pi, np.pi, n)

# Compute Y = sin(2X), which gives a sine wave with double frequency
Y = np.sin(2 * X)

# Manually define the axes position on the figure: (left, bottom, width, height)
# All values are relative (0 to 1), making the plot fill nearly the entire figure
plt.axes((0.025, 0.025, 0.95, 0.95))

# Plot the curve Y + 1 (shifted upward) in solid blue with full opacity
plt.plot(X, Y + 1, color="blue", alpha=1.00)

# Fill the area between the curve Y + 1 and the line y = 1
# This shades the area under the blue curve with partial transparency
plt.fill_between(X, 1, Y + 1, color="blue", alpha=0.25)

# Plot the curve Y - 1 (shifted downward) in solid blue
plt.plot(X, Y - 1, color="blue", alpha=1.00)

# Fill the area between the curve Y - 1 and the line y = -1
# Condition: where (Y - 1) > -1  this fills in blue when the curve is above -1
plt.fill_between(X, -1, Y - 1, (Y - 1) > -1, color="blue", alpha=0.25)

# Fill the area where (Y - 1) < -1  this fills in red when the curve is below -1
plt.fill_between(X, -1, Y - 1, (Y - 1) < -1, color="red", alpha=0.25)

# Set limits for the x-axis and remove tick marks
plt.xlim(-np.pi, np.pi)
plt.xticks([])

# Set limits for the y-axis and remove tick marks
plt.ylim(-2.5, 2.5)
plt.yticks([])

# Display the final plot
plt.show()
```

**Histogram**

The histogram represents the distribution of a numerical variable by dividing the range of data into intervals and showing how many data fall into each interval (for continuous variables) or the frequency of each category (for categorical variables).

```python
# Generate 1000 random samples from a standard normal distribution (mean = 0, std = 1)
data = np.random.randn(1000)

# Create a new figure with size 10 inches wide by 5 inches tall
plt.figure(figsize=(10, 5))

# Create a histogram of the data:
# - `bins=30`: divide the data range into 30 intervals (bars)
# - `alpha=0.7`: set transparency level (1.0 is opaque, 0 is fully transparent)
plt.hist(data, bins=30, alpha=0.7)

# Add a title to the plot
plt.title("Histogram")

# Display the final plot
plt.show()
```

**Bar chart**

The bar plot represents categorical data with rectangular bars with heights (or lengths, in the case of horizontal bars) proportional to the values they represent.

```python
# Define the category labels for each bar
labels = ["A", "B", "C", "D"]

# Define the numeric values corresponding to each label
values = [10, 20, 15, 30]

# Create a new figure with size 10 inches wide by 5 inches tall
plt.figure(figsize=(10, 5))

# Create a vertical bar chart:
# - `labels` defines the categories on the x-axis
# - `values` defines the height of each bar
plt.bar(labels, values)

# Add a title to the plot
plt.title("Bar chart")

# Display the plot
plt.show()
```

**Pie chart**

A pie chart represents data in circular sectors, where each sector corresponds to a category and its size is proportional to the value it represents.

```python
# Define category labels for each slice of the pie
labels = ["A", "B", "C", "D"]

# Define the size (proportion) of each slice  the values dont need to sum to 100
sizes = [215, 130, 245, 210]

# Create a new figure with a square size to keep the pie circular
plt.figure(figsize=(7, 7))

# Add a title to the chart
plt.title("Pie chart")


explode = [0, 0, 0.1, 0]  # Explode third slice (C), 0.1 = 10% offset from center
colors = ['skyblue', 'lightgreen', 'salmon', 'gold'] #Add custom colors
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', explode=explode)#Add percentage labels

# Display the pie chart
plt.show()
```

**Boxplot**

A boxplot shows the distribution of quantitative data by its quartiles and possibly outliers.

Screenshot 2025-07-15 204508.png

The ends of the box indicate the lower and upper quartiles, while the line inside the box indicates the median.

```python
# Example data for a boxplot
data = np.random.normal(0, 1, 100)  # Generate 100 data points from a standard normal distribution

# Create a new figure with size 10 inches wide by 5 inches tall
plt.figure(figsize=(10, 5))

# Create a boxplot of the data:
# - The box represents the interquartile range (IQR)
# - The line inside the box is the median
# - The "whiskers" extend to 1.5 times the IQR from the box ends or the minimum/maximum data points (whichever is smaller)
# - Points outside the whiskers are plotted as outliers
plt.boxplot(data)

# Add a title to the plot
plt.title("Boxplot Example")

# Display the plot
plt.show()
```

**As you see there are lots of possibilities, it's obvious we can't cover every plot in a detailed way. The thing is you can discover more plots and ways to transmit your results, so go for it. We let you with this example for you to indague:**

```python
# Define a 2D scalar function of x and y
# This function combines polynomial and exponential decay to create interesting surface patterns
def f(x, y):
    return (1 - x / 2 + x**5 + y**3) * np.exp(-(x**2) - y**2)

# Set the resolution of the grid
n = 256

# Create 1D arrays for x and y values ranging from -3 to 3
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)

# Create a 2D grid of coordinates from the x and y arrays
X, Y = np.meshgrid(x, y)  # X and Y are 2D arrays of shape (n, n)

# Manually set the position and size of the axes inside the figure
# (left, bottom, width, height), all in [0, 1] relative coordinates
plt.axes((0.025, 0.025, 0.95, 0.95))

# Plot a filled contour plot (smooth gradient regions)
# - `8` is the number of contour levels
# - `alpha=0.75` controls the transparency
# - `cmap="hot"` uses the 'hot' color map (black-red-orange-yellow)
plt.contourf(X, Y, f(X, Y), 8, alpha=0.75, cmap="hot")

# Overlay contour lines on top of the filled contour
# - Uses same function values
# - 8 contour levels
# - Black contour lines with width 0.5
C = plt.contour(X, Y, f(X, Y), 8, colors="black", linewidths=0.5)

# Add labels to the contour lines
# - `inline=1` draws labels into the lines
# - `fontsize=10` sets the label size
plt.clabel(C, inline=1, fontsize=10)

# Remove x and y axis tick marks for a clean plot
plt.xticks([])
plt.yticks([])

plt.title("Contour Example")

# Display the final figure
plt.show()
```

**Exercise 05: Starting from the code below, try to reproduce the graphic by adding labels for red bars. Hint You need to take care of text alignment ().**

Screenshot 2025-07-15 211745.png

```python
n = 12
X = np.arange(n)
rng = np.random.default_rng()
Y1 = (1 - X / float(n)) * rng.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * rng.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')

plt.ylim(-1.25, +1.25)
```
